```python
basePath = "RawData/exp05/"
cnt = 0
res = []
for dir in os.listdir(basePath):
    if ("meta.log" == dir):
        continue
    dirname = basePath+dir
    #print(dirname)
    image = np.load(dirname+"/"+os.listdir(dirname)[0])
    if (dir == "realsense_color"):
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image)
    if (dir in ["MLX", "seek_thermal", "senxor_m08", "senxor_m08_1"]):
        res.append(image)
        cbar = plt.colorbar()
        cbar.set_label('Temperature')
        #plt.axis('off')
    plt.title(dir)
    plt.show()
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(2, 2, figsize = (10, 10))

im0 = ax[0,0].imshow(res[0])
ax[0, 0].set_title("MLX")
cbar0 = fig.colorbar(im0, ax=ax[0, 0])
cbar0.set_label('Temperature')

im1 = ax[0,1].imshow(res[1])
ax[0,1].set_title("seek")
cbar1 = fig.colorbar(im1, ax=ax[0, 1])
cbar1.set_label('Temperature')

im2 = ax[1,0].imshow(res[2])
ax[1, 0].set_title("senxor_m16")
cbar2 = fig.colorbar(im2, ax=ax[1, 0])
cbar2.set_label('Temperature')

im3 = ax[1,1].imshow(res[3])
ax[1,1].set_title("senxor_m08")
cbar3 = fig.colorbar(im3, ax=ax[1, 1])
cbar3.set_label('Temperature')

plt.show()

```
